```python
import statistics
import random
import matplotlib.pyplot as plt
from statistics import mode
```

```python
def lanzar(iteraciones):
    lista_nums = [0] * iteraciones
    for i in range(iteraciones):
        lista_nums[i] = random.randint(1,6)
    return lista_nums
```

```python
list_of_100 = lanzar(100)
prom_of_100 = sum(list_of_100) / len(list_of_100)
print("El promedio con 100 lanzamientos es: " , prom_of_100 )
print("El número más frecuente fue el: ", mode(list_of_100))

plt.hist(list_of_100, edgecolor="white")
plt.xlabel('Suma')
plt.ylabel('# de repeticiones')
plt.title('Resultados: ')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Simulamos lanzamientos de un dado justo
np.random.seed(42)  # Para reproducibilidad
num_lanzamientos = np.arange(1, 1001)
resultados = np.random.randint(1, 7, 1000)
frecuencia_relativa = np.cumsum(resultados == 6) / num_lanzamientos

# Graficamos la convergencia
plt.figure(figsize=(10, 6))
plt.plot(num_lanzamientos, frecuencia_relativa, label='Frecuencia Relativa de 6')
plt.axhline(y=1/6, color='r', linestyle='--', label='Probabilidad Teórica (1/6)')
plt.title('Convergencia de la Ley de los Grandes Números')
plt.xlabel('Número de Lanzamientos')
plt.ylabel('Frecuencia Relativa')
plt.legend()
plt.show()
```
