## Método Aceptar-Rechazar 

El método de aceptación-rechazo (o accept-reject) en Monte Carlo es una técnica utilizada para generar números aleatorios a partir de una distribución de probabilidad arbitraria. A continuación, te presento un ejemplo simple de un programa en Python que implementa el método aceptar-rechazar:

```python
import numpy as np
import matplotlib.pyplot as plt
```

La función 'funcion_objetivo' representa la función de probabilidad que deseas muestrear. La función 'generador_aleatorio' genera un par de números aleatorios (x, y) en el rango de interés. La función aceptar_rechazar realiza el método aceptar-rechazar y devuelve las muestras aceptadas.

```python
def funcion_objetivo(x, funcion):
    #Se define la función que se quiere probar. 
    if funcion == 1:
        return np.exp(-x**2) + 0.5 * np.sin(4*x)
    else:
        return np.exp(-x)
    
def generador_aleatorio():
    # Genera un número aleatorio uniforme en el rango de interés
    x = np.random.uniform(0, 2, 1)[0]
    # Genera un número aleatorio entre 0 y el valor máximo de la función objetivo
    y = np.random.uniform(0, 1, 1)[0] * 1.5
    return x, y
```

```python
def aceptar_rechazar(iteraciones):
    muestras_aceptadas = []

    for _ in range(iteraciones):
        x, y = generador_aleatorio()
        if y <= funcion_objetivo(x, 1):
            muestras_aceptadas.append(x)

    return muestras_aceptadas


# Número de iteraciones
num_iteraciones = 10000

# Generar muestras usando el método aceptar-rechazar
muestras_aceptadas = aceptar_rechazar(num_iteraciones)
```

### Se compara el histograma de las muestras aceptadas con la forma de la función objetivo. 

```python
# Visualizar los resultados
plt.hist(muestras_aceptadas, bins=50, density=True, alpha=0.7, color='blue', label='Muestras Aceptadas')
x = np.linspace(0, 2, 100)
plt.plot(x, funcion_objetivo(x, 1), 'r', linewidth=2, label='Función Objetivo')
plt.legend()
plt.title('Método Aceptar-Rechazar en Monte Carlo')
plt.xlabel('Variable Aleatoria')
plt.ylabel('Densidad de Probabilidad')
plt.show()
```
