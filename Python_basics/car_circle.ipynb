```python
from matplotlib import pyplot as plt
from IPython import display
import math
import numpy as np
```

```python
epoca = 1
#Car 1
x_car1 = 0
y_car1 = 0

#Car 2
x_car2 = 0
y_car2 = 0

theta_1 = 0
theta_2 = 0

#Circle 1
x_circle1 = np.linspace( -0.9 , 0.9 , 250 )
y_circle1 = np.linspace( -0.9 , 0.9 , 250 )
a_circle1, b_circle1 = np.meshgrid( x_circle1 , y_circle1 )
C_circle1 = a_circle1 ** 2 + b_circle1 ** 2 - 0.8
```

```python
for i in range(200):
  x_car1 = math.cos(theta_1)
  y_car1 = math.sin(theta_1)
  
  x_car2 = math.cos(theta_2)
  y_car2 = math.sin(theta_2)
  
  if (abs(x_car1 - x_car2) < 0.3 and abs(y_car1 - y_car2) < 0.3):
    x_car1 = 0.8*x_car1
    y_car1 = 0.8*y_car1
  if (abs(x_car1 - x_car2) < 0.1 and abs(y_car1 - y_car2) < 0.1):
    x_car1 = 0.6*x_car1
    y_car1 = 0.6*y_car1
    
  
  plt.figure(figsize=(7,7))
  plt.grid()
  plt.contour( a_circle1 , b_circle1 , C_circle1 , [0] )
  
  plt.xlim(-2,2)
  plt.ylim(-2,2)
  
  plt.plot(x_car1, y_car1)
  plt.scatter(x_car1, y_car1, label = 'X  car 1: ' + str(x_car1) + ' Y car 1: '+ str(y_car1), s=200)
  
  plt.plot(x_car2, y_car2)
  plt.scatter(x_car2, y_car2, label = 'X  car 2: ' + str(x_car2) + ' Y car 2: '+ str(y_car2), s=200)
    
  plt.legend(loc="upper left")
  display.clear_output(wait=True)
  plt.show()
    
  epoca += 1
  theta_1 += 0.20
  theta_2 += 0.10

```
