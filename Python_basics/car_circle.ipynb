```python
def pedir_centro():
    centro_x = int(input("Ingresa el punto x para el centro de la pista: "))
    centro_y = int(input("Ingresa el punto y para el centro de la pista: "))
    return centro_x, centro_y
```

```python
def pedir_diametro():
    ban = True
    while ban == True:
        r_car1 = int(input("Ingresa el diámetro del carril del primer carro: "))
        r_car2 = int(input("Ingresa el diámetro del carril del segundo carro: "))
        if r_car1 == r_car2:
            print("Los diámetro no pueden ser iguales, intenta de nuevo.")
        else:
            ban = False
    return r_car1, r_car2
```
