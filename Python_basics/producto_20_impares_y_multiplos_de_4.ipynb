```python
impares = []
multiplos_4 = []
productos= []

def funcion_producto():
  for i in range(1,41,2):
    impares.append(i)
  for i in range(1,21):
    multiplos_4.append(4*i)
  for i in range(0,19):
    productos.append(impares[i]*multiplos_4[i])
  return productos
```

```python
print(funcion_producto())
```
