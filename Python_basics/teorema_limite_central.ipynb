## Teorema de límite central

```python
def exponential_function(x):
    return ((2*x) ** 2) + 2
    
```

```python
num_muestras = int(input('Ingresa el número de muestras a generar'))
num_muestras
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 100, num_muestras)
y = exponential_function(x)

plt.figure(figsize=(7,7))
plt.plot(x, y)
plt.show()
```
