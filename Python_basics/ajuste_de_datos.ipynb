## Ajuste de datos

```python
import numpy as np
import matplotlib.pyplot as plt
```

Generación de lambda

```python
# Establecer el parámetro lambda
lambda_param = 0.5

# Generar 10,000 puntos distribuidos exponencialmente
datos_exponenciales = np.random.exponential(scale=1/lambda_param, size=100)


# Estimador de máxima verosimilitud para lambda
lambda_estimado = 1 / np.mean(datos_exponenciales)

print(f"Valor real de lambda: {lambda_param}")
print(f"Estimador de máxima verosimilitud para lambda: {lambda_estimado}")
```

Se utiliza el valor generado de lambda obtenido anteriormente para 100 números

```python
datos_exponenciales = np.random.exponential(scale=1/lambda_estimado, size=100)

# Visualizar el histograma de los datos generados
plt.hist(datos_exponenciales, bins=50, density=True, alpha=0.7, color='blue', label='Datos Exponenciales')
plt.title(f'Distribución Exponencial (lambda={lambda_estimado})')
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.show()
```
