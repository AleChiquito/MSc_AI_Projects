```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def golden_section_search(f, a, b, tol=1e-6, find_maximum=False):
    golden_ratio = (1 + math.sqrt(5)) / 2
    inv_golden_ratio = 1 / golden_ratio

    x1 = b - inv_golden_ratio * (b - a)
    x2 = a + inv_golden_ratio * (b - a)

    f1 = f(x1)
    f2 = f(x2)

    while (b - a) > tol:
        if find_maximum:
            if f1 < f2:
                b = x2
                x2 = x1
                f2 = f1
                x1 = b - inv_golden_ratio * (b - a)
                f1 = f(x1)
            else:
                a = x1
                x1 = x2
                f1 = f2
                x2 = a + inv_golden_ratio * (b - a)
                f2 = f(x2)
        else:
            if f1 > f2:
                b = x2
                x2 = x1
                f2 = f1
                x1 = b - inv_golden_ratio * (b - a)
                f1 = f(x1)
            else:
                a = x1
                x1 = x2
                f1 = f2
                x2 = a + inv_golden_ratio * (b - a)
                f2 = f(x2)

    xmin = (a + b) / 2
    fmin = f(xmin)
    return xmin, fmin
```

```python
# Ejemplo de uso
def objective_function(x):
    return x**2 + 4 * np.sin(x)
```

```python
# Intervalo inicial de búsqueda
a, b = -2, 4

# Llamada al algoritmo de Búsqueda de la Sección Dorada para encontrar el mínimo
xmin, fmin = golden_section_search(objective_function, a, b, 1e-6)

# Llamada al algoritmo de Búsqueda de la Sección Dorada para encontrar el máximo
xmax, fmax = golden_section_search(objective_function, a, b, 1e-6, find_maximum=True)
```

```python
# Graficar la función y los extremos encontrados
x_values = np.linspace(-3, 4, 100)
y_values = objective_function(x_values)

plt.plot(x_values, y_values, label='Función objetivo')
plt.scatter([xmin, xmax], [fmin, fmax], color='red', label='Extremos encontrados')
plt.title('Gráfica de la función y extremos encontrados')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
