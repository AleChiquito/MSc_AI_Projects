## Interpolacion polinomial

```python
import math as m
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=[0.9,1.3,1.9,2.1,2.6,3.0,3.9,4.4,4.7,5.0,6.0,7.0,8.0,9.2,10.5,11.3,11.6,12.0,12.6,13.0,13.3];
f_X=[1.3,1.5,1.85,2.1,2.6,2.7,2.4,2.15,2.05,2.1,2.25,2.3,2.25,1.95,1.4,0.9,0.7,0.6,0.5,0.4,0.25];
```

```python
#Se define el grado de interpolaci√≥n
grado = len(X) - 1

#Se calcuclan las dierencias cuadradas
diferencias = f_X
b = [f_X[0]]

for k in range(1,grado+1):
    D = []
    for i in range(len(diferencias)-1):
        dif = (diferencias[i+1] - diferencias[i]) / (X[i+k]-X[i])
        D.append(dif)
    b.append(D[0])#Se guardan los valores de b
    diferencias = D
```

```python
F_final = []
xi = np.arange(min(X), max(X)+0.1, 0.1)
xi = xi.tolist()

for xr in xi:
    f = b[0]#para calcular la suma total, el primer valor es b0
    cuentas = []#variable auxiliar

    for i in range(1,grado+1):#dependiendo del grado del elemento al que se aplica formula
        cuenta = 1#cuenta se inicializa con el neutro multiplicador.
        for j in range(i):
            c = xr - X[j]
            cuenta = cuenta * c#se multiplica lo necesario((x ‚Äì x0)*(x ‚Äì x1)*...(x - xn))
        cuentas.append(cuenta)

    for i in range(grado):
        a = b[i+1] * cuentas[i]
        print(a)
        f = f + a
    F_final.append(f)
```

```python
#Graficando
fig, ax = plt.subplots()
ax.plot(xi, F_final,color='blue',label = 'Funci√≥n polinomail grado 20')
ax.plot(X, f_X,color='m',marker ="o", label = 'Punto a estimar')
ax.plot(X, f_X,color='c', label = 'Linea Original')
ax.set_title(f'Interpolaci√≥n Polinomial de grado:{grado}', loc = "center", fontdict = {'fontsize':9, 'fontweight':'bold', 'color':'black'})
ax.set_xlabel("x", fontdict = {'fontsize':9, 'fontweight':'bold', 'color':'black'})
ax.set_ylabel("y", fontdict = {'fontsize':9, 'fontweight':'bold', 'color':'black'})
ax.grid()
ax.legend(loc = 'upper left',fontsize=9)
plt.show()
```
