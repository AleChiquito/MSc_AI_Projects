## Regresión Cuadrática

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Declaración de puntos
#x=[1, 2, 3, 4, 5, 6, 7]
#y=[0.5, 2.5, 2.0, 4.0, 3.5, 6.0, 5.5]

x=[0,1,2,3]
y=[1,3,2,0]
y = np.array(y)
y = np.array(y)
```

```python
#Función para elevar al cuadrado una matriz
def lista_al_cuadrado(lista):
    lista2 = []
    for i in range(0,len(lista)):
        lista2.append(lista[i]*lista[i])
    return lista2

#Función para elevar al cubo una matriz
def lista_al_cubo(lista):
    lista3 = []
    for i in range(0,len(lista)):
        lista3.append(lista[i]*lista[i]*lista[i])
    return lista3

#Función para elevar a la cuarta a una matriz
def lista_cuarta(lista):
    lista4 = []
    for i in range(0,len(lista)):
        lista4.append(lista[i]*lista[i]*lista[i]*lista[i])
    return lista4

#Función para multiplicar matrices
def lista_dot(lista1,lista2):
    Lista_multi = []
    for i in range(0,len(lista1)):
        Lista_multi.append(lista1[i]*lista2[i])
    return Lista_multi
```

```python
# Elementos de la matriz
n=len(x);

#Se crean listas con las sumas y potencias de las matrices
sum_xi=sum(x)
sum_xi2=sum(lista_al_cuadrado(x))
sum_xi3=sum(lista_al_cubo(x))
sum_xi4=sum(lista_cuarta(x))
```

```python
sum_yi=sum(y)
sum_xiyi=sum(lista_dot(x,y))
sum_xi2yi=sum(lista_dot(lista_al_cuadrado(x),y))
```

```python
# Se crean una matriz con los resultados obtenidos anteriormente
A = [[n,      sum_xi, sum_xi2],
    [sum_xi, sum_xi2, sum_xi3],
    [sum_xi2, sum_xi3, sum_xi4]]
```

```python
# Se crean una matriz con las sumas obtenidos anteriormente
d = [[sum_yi],
    [sum_xiyi],
    [sum_xi2yi]]
d = np.array(d)
```

```python
#Operación entre matrices declaradas anteriormente
b = np.linalg.inv(A).dot(d)
```

```python
# Se obtienen los coeficientes para la ecuación: 
#Y = ax^2 + bx + c
x0 = b[0]
x1 = b[1]
x2 = b[2]
```

```python
# Se aplica la ecuación Y = ax^2 + bx + c
def y_regresion(x,a0,a1,a2):
    return a0 + (a1*x) + a2*lista_al_cuadrado(x)
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y,color='blue',marker ="o",label = f'Puntos')
ax.plot(x, y_regresion(x,x0,x1,x2), 'red',label =f'Regresión cuadrática') #R^2:{round(r_2,3)} ; r:{round(r,3)}')
#ax.plot(label = f'R^2:{r_2}')
ax.set_title('Regresión Cuadratica', loc = "center", fontdict = {'fontsize':9, 'fontweight':'bold', 'color':'black'})
ax.set_xlabel("X", fontdict = {'fontsize':9, 'fontweight':'bold', 'color':'black'})
ax.set_ylabel("Y", fontdict = {'fontsize':9, 'fontweight':'bold', 'color':'black'})
ax.grid()
ax.legend(loc = 'upper left')
plt.show()
```
