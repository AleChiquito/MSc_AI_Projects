```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Función definida para analizar
def f(x):
    return np.sin(x)+5*x
```

```python
# Se genera una lista de 100 puntos para evaluar la función
x = np.linspace(-4, 8, 100)
print(x)

#Se evalua la función 
y = f(x)
```

```python
def biseccion(f,lim_inf,lim_sup,iteraciones,tol=1e-5):
  tabla=pd.DataFrame(columns=['lim_inf','lim_sup','lim_med','punto_inf','punto_sup','punto_med','Error(%)'])

  for idx in range(iteraciones):
    lim_med = (lim_inf + lim_sup) / 2.0
    if idx == 0:
      c_1 = 0
      error = 0

    tabla=tabla._append({'lim_inf':lim_inf,'lim_sup':lim_sup,'lim_med':lim_med,'punto_inf':f(lim_inf),'punto_sup':f(lim_sup),'punto_med':f(lim_med),'Error(%)':error*100}, ignore_index=True)

    if f(lim_med) == 0:
        break
    elif f(lim_inf) * f(lim_med) < 0:
        lim_sup = lim_med
    else:
        lim_inf = lim_med
    c_1 = (lim_inf + lim_sup) / 2.0
    error = abs((c_1-lim_med)/c_1)
    if error < tol:
       break

  return lim_med, tabla
```

```python
raiz,tabla = biseccion(f,-1,2,100)
print("Raiz: ",raiz)
print(tabla)
```

```python
plt.title('Función: sin(x) + 5x')
plt.plot(x, y, lw=2)
plt.scatter(raiz, 0, color='red')
plt.xlim(-1,1)
plt.ylim(-2,4)
plt.grid(True, linestyle='-.')

plt.show()
```
