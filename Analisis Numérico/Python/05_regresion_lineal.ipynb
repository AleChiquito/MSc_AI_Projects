## Minimos Cuadrados

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Creando array de 0 a 6 en paos de 0.1
x = np.arange(1, 8, 1)
#Creando array
y = np.array((0.5,2.5,2,4,3.5,6,5.5))

#Se crean variables con sumas de los arrays
n = np.size(x)
sum_xi = np.sum(x)
sum_xi2 = np.sum(np.power(x,2))
sum_yi = np.sum(y)
sum_xiyi = np.sum(np.multiply(x, y))
```

```python
#Creando matriz de 2X2 con las sumas creadas anteriormente
matrix_A = np.array([[n,sum_xi], [sum_xi,sum_xi2]])
matrix_A
```

```python
#Creando matriz de 2X1 con las suma de Y y XY
matrix_B = np.array([[sum_yi],[sum_xiyi]])
matrix_B
```

```python
#Operación entre matrices
result_matrix = np.dot(np.linalg.inv(matrix_A),matrix_B)
result_matrix
```

```python
x0 = result_matrix[0]
x1 = result_matrix[1]
print(x0, x1)
```

```python
# Se crea la aproximación utilizando la ecuación de la recta
x_aprox = x
y_aprox = x0 + (x1 * x_aprox)
print(x_aprox, y_aprox)
```

```python
#Se grafican los puntos y la recta creada anteriormente
fig, ax = plt.subplots()
ax.scatter(x, y,color='blue',marker ="o",label = 'Puntos')
ax.plot(x_aprox, y_aprox,color='orange', label = 'Regresión Lineal')
ax.set_title('Regresión Lineal Minimos Cuadrados', loc = "center", fontdict = {'fontsize':9, 'fontweight':'bold', 'color':'black'})
ax.set_xlabel("X", fontdict = {'fontsize':9, 'fontweight':'bold', 'color':'black'})
ax.set_ylabel("Y", fontdict = {'fontsize':9, 'fontweight':'bold', 'color':'black'})
ax.grid()
ax.legend(loc = 'upper left')
plt.show()
```
