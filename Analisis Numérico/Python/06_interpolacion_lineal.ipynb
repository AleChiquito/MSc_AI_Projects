## Interpolación Lineal

Estime el logaritmo natural de 2 mediante interpolación lineal. Primero, realice el cálculo por interpolación entre ln 1 = 0 y ln 6 = 1.791759. Después, repita el procedimiento, pero use un intervalo menor de ln 1 a ln 4 (1.386294). Observe que el valor verdadero de ln 2 es 0.6931472.

```python
import math as m
import numpy as np
import matplotlib.pyplot as plt
```

```python
x0 = 1
x1 = 6

f_x0 = m.log(x0)
f_x1 = m.log(x1)
```

```python
logs = []#Lista para los logaritmos naturales
#Se crea un array de 0 a 6 con separaciones de 0.1
x_values = np.arange(x0, x1+0.1, 0.1)
```

```python
for j in range(np.size(x_values)):
    x_values[j] = round(x_values[j],3)
    #Se guarda los logaritmos de cada elemento del array en una lista
    logs.append(m.log(round(x_values[j],3)))
```

```python
#Valor a estimar
x = 2
fx_ver = m.log(x)
```

```python
#Se definen listas para guardar los puntos en cada iteración.
aproximaciones ,X_puntos , Y_puntos, aux  = [],[],[],[]
```

```python
for i in range(6,1,-1):
    plt.clf()
    f_x1 = m.log(i)
    # Ecuación de interpolación lineal
    f_x = (f_x0) + ((f_x1-f_x0)/(i-x0))*(x-x0)
    aproximaciones.append(f_x)
    aux.append(x)
    X = [x0,i]
    Y = [f_x0,f_x1]
    X_puntos.append(X)
    Y_puntos.append(Y)
    print(X,Y)
    plt.scatter(x, fx_ver,color='orange',marker ="o",s=10**2,label = 'Valor real')
    plt.plot(x_values, logs,color='red', label = 'Grafica real')
    plt.scatter(aux,aproximaciones,color='black',marker ="o",label = 'Aproximación')
    for k in range(len(X_puntos)):
        plt.plot(X_puntos[k],Y_puntos[k],color='green')
    plt.title('Interpolacion Lineal de ln(2)', loc = "center")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid()
    plt.legend(loc = 'upper left')
    plt.tight_layout()
    plt.pause(2)
    plt.ioff()
plt.show()
```
