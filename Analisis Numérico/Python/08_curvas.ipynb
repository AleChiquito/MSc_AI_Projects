## Curvas Parametricas

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xi= [-1, 0, 1, 0, 1];
yi= [0, 1, 0.5, 0, -1];
values = np.arange(0, 1.01, 0.01)

x_t = (((64*values-(352/3))*values+60)*values-(14/3))*values-1
y_t = ((((-64/3)*values+48)*values-(116/3))*values+11)*values
```

```python
fig0, ax = plt.subplots()
ax.scatter(xi, yi,color='green', marker='o',label = 'Puntos reales')
ax.plot(xi, yi,color='blue', label = 'Aproximaciones')
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.grid(color = 'gray', linestyle = 'dashed')
ax.legend(loc = 'upper left', fontsize=10)
```

```python
fig1 = plt.figure(figsize=(6,6))
fig1.tight_layout()
ax0 = plt.subplot(1,2,1)
ax0.plot(values,x_t)
ax0.set_xlabel('t')
ax0.set_ylabel('x(t)')
ax0.set_title('Curva de x con respecto a t')
ax0.grid()


ax1 = plt.subplot(1,2,2)
ax1.plot(values,y_t)
ax1.set_xlabel('t')
ax1.set_ylabel('y(t)')
ax1.set_title('Curva parametrica de y con respecto a t')
ax1.grid()
```

```python
fig2, ax = plt.subplots()
ax.scatter(xi, yi,color='green')
ax.plot(x_t, y_t,color='blue')
ax.set_title('Curva de x(t) y y(t)', loc = "center")
ax.set_xlabel("x(t)")
ax.set_ylabel("y(t)")
ax.grid()
plt.show()
```
