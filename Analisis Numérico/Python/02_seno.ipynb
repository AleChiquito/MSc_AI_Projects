# Función $sen(x)$

$senx = x - x^3/3! + x^5/5! - x^7/7! + x^9/9! - ...$

```python
import math

resultados = []
def seno(x, lim):
    sum = 0
    for i in range(0,lim):
        resultado = math.pow(-1,i) / math.factorial(2 * i + 1) * math.pow(x,2 * i + 1)
        sum += resultado
        resultados.append(resultado)
    return sum
```

```python
lim = int(input("Ingresa el número límite de la serie: "))
x = float(input("Ingresa el valor de x: "))
#Se convierte a radianes
x = x*math.pi/180

Valor_aprox = seno(x, lim)
Valor_aprox
```

```python
Valor_verd = math.sin(x)
Valor_verd
```

### Error verdadero

```python
error_verd = Valor_verd - Valor_aprox
print("El error verdadero es: ", error_verd)
```

### Error relativo porcentual verdadero

```python
error_rel = ((Valor_verd - Valor_aprox) / Valor_verd ) * 100
print("El Error relativo porcentual verdadero es: ", error_rel, " %")
```

```python
import matplotlib.pyplot as plt


def create_graph(results, lim, titulo):
    plt.plot(results, range(0, lim), color='red', marker='o')
    plt.title(titulo)
    plt.xlabel("Resultado por termino")
    plt.ylabel("# de termino")
    plt.show()
```

```python
create_graph(resultados, lim, "Resultados por términos de función Seno")
```
