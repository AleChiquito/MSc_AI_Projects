## Interpolaci√≥n Segmentaria Lineal - spline 

```python
import math as m
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
#Se definen los puntos, x y y
x = [3.0, 4.5, 7.0, 9.0]
f_x = [2.5, 1.0, 2.5, 0.5]

leng = len(x)
f1x,f1x_=[],[]
x_ = []
```

```python
for i in range(leng-1):
    plt.clf()
    #Se calcula la pendiente entre dos puntos
    m = (f_x[i+1]-f_x[i]) / (x[i+1]-x[i])
    #Se genera un array de los dos puntos adyacentes
    x = np.arange(x[i], x[i+1]+0.01, 0.01)
    f1x = f_x[i] + m*(x-x[i])
    f1x_ = np.append(f1x_,f1x)
    x_= np.append(x_,x)
    plt.scatter(x,f_x,color='blue',marker='o')
    #Se gr√°fica la recta de cada iteraci√≥n
    for k in range(leng-2):
        plt.plot(x_,f1x_,color='red')
        plt.title('Interpolaci√≥n Segmentaria Lineal')
        plt.xlim(0,14)
        plt.ylim(-2,8)
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.grid(True)
    plt.tight_layout()
    plt.pause(1)
    plt.ioff()

plt.show()
```
