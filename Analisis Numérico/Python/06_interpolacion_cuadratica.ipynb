## Interpolación Cuadráctica

```python
import math as m
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Puntos de partida
x0 = 1
x1 = 4
x2 = 6

#logaritmos de los puntos de partida
log_x0 = m.log(x0)
log_x1 = m.log(x1)
log_x2 = m.log(x2)
```

```python
#Lista para los logaritmos naturales
f_logs = []
#Creando array del punto 0 hasta 10 en pasos de 0.1
xi = np.arange(x0, 10.1, 0.1)
```

```python
for j in range(np.size(xi)):
    xi[j] = round(xi[j],3)
    #Se guardan todos los logs 
    f_logs.append(m.log(round(xi[j],3)))


#Valor a estimar
x = 2
#Valor real
f1_ver = m.log(x)
```

```python
#Primera diferencia dividida
f_x1_x0 = (log_x1 - log_x0)/ (x1-x0)
f_x2_x1 = (log_x2 - log_x1)/ (x2-x1)

#Segunda diferencia divididas
f_x2_x1_x0 = (f_x2_x1-f_x1_x0)/(x2-x0)

#Coeficientes de b
b0 = log_x0
b1 = f_x1_x0
b2 = (((log_x2 - log_x1)/(x2-x1))-((log_x1 - log_x0)/(x1-x0)))/(x2-x0)

#Coeficientes par simplicar la ecuacion
a0 = b0 - b1*x0 + b2*x0*x1
a1 = b1-b2*x0-b2*x1
a2 = b2
```

```python
xi_ref = np.arange(1, 10.01, 0.01)
f_est = a0 + np.multiply(xi_ref,a1) + np.multiply(xi_ref**2,a2)
```

```python
#Generando el punto de aproximación
x_sqr = 2;
f_sqr =  a0 + np.multiply(x_sqr,a1) + np.multiply(x_sqr**2,a2)
```

```python
fig, ax = plt.subplots()
ax.plot(xi, f_logs,color='blue',label = 'Función Log')
ax.plot(x, f1_ver,color='red',marker ="o", label = 'Punto a estimar')
ax.plot(x0, log_x0,color='b',marker ="o", label = 'Puntos dados',)
ax.plot(x1, log_x1,color='b',marker ="o")
ax.plot(x2, log_x2,color='b',marker ="o")
ax.plot(xi_ref,f_est,color='cyan',label = 'Función de aproximación')
ax.plot(x_sqr,f_sqr,color='cyan',marker ="o",label = 'Punto de aproximación')

ax.set_title('Interpolación Cuadratica de ln(x)', loc = "center")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.grid()
ax.set_xlim(xi[0],xi[-1])
ax.legend(loc = 'upper left',fontsize=9)
plt.show()
```
