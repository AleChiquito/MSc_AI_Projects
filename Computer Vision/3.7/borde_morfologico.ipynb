```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
ImgOrg = cv2.imread('celulas.png', cv2.IMREAD_GRAYSCALE)

if len(ImgOrg.shape) == 3:
    ImgOrg = cv2.cvtColor(ImgOrg, cv2.COLOR_BGR2GRAY)
```

```python
_, ImgOrgBin = cv2.threshold(ImgOrg, 128, 255, cv2.THRESH_BINARY)

# Definir el elemento estructurante
str1 = np.ones((3, 3), np.uint8)

# Dilatación
ImgDil = cv2.dilate(ImgOrgBin, str1)

# Borde Morfológico
ImgFinal = cv2.subtract(ImgDil, ImgOrgBin)

# Esqueleto
ImgSkel = cv2.ximgproc.thinning(ImgOrgBin, thinningType=cv2.ximgproc.THINNING_ZHANGSUEN)
```

```python
# Mostrar imágenes
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(ImgOrg, cmap='gray')
plt.title("Imagen Original (B)")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(ImgFinal, cmap='gray')
plt.title("Borde Morfológico [(B + E) - B]")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(ImgSkel, cmap='gray')
plt.title("Skeleton")
plt.axis('off')

plt.show()
```
