```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
examinar1 = 'egipt.jpg'
A = cv2.imread(examinar1)
A = cv2.cvtColor(A, cv2.COLOR_BGR2RGB)
A = A.astype(np.float32) / 255.0
```

```python
# Kernels de convolucin
k55 = np.ones((5, 5)) / 25.0
k11 = np.ones((11, 11)) / 121.0
k21 = np.ones((21, 21)) / 441.0
k5 = np.ones((1, 5)) / 5.0
k6 = np.ones((3, 1)) / 3.0

# Aplicar convolucin
R1 = cv2.filter2D(A, -1, k55)
R2 = cv2.filter2D(A, -1, k11)
R3 = cv2.filter2D(A, -1, k21)
R4 = cv2.filter2D(A, -1, k5)
R5 = cv2.filter2D(A, -1, k6)

# Mostrar imgenes
plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
plt.imshow(A)
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(R1)
plt.title('Mscara 5x5')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(R2)
plt.title('Mscara 11x11')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(R3)
plt.title('Mscara 21x21')
plt.axis('off')
```

```python
plt.figure(figsize=(16, 12))

plt.subplot(2, 2, 1)
plt.imshow(R4)
plt.title('Media Horizontal')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(R5)
plt.title('Media Vertical')
plt.axis('off')

plt.show()
```
