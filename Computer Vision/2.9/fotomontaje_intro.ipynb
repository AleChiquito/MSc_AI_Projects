```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = cv2.imread('iglesia_color.jpg')
B = cv2.imread('chico_color.jpg')

# Las images deben tener el mismo tamaño
B = cv2.resize(B, (A.shape[1], A.shape[0]))
```

```python
# Convertir las imágenes a escala de grises
A_gray = cv2.cvtColor(A, cv2.COLOR_BGR2GRAY)
B_gray = cv2.cvtColor(B, cv2.COLOR_BGR2GRAY)

# Aplicar las operaciones lógicas a las imágenes originales
and_result = cv2.bitwise_and(A, B)
or_result = cv2.bitwise_or(A, B)
xor_result = cv2.bitwise_xor(A, B)

# Aplicar las operaciones lógicas a las imágenes en escala de grises
and_result_gray = cv2.bitwise_and(A_gray, B_gray)
or_result_gray = cv2.bitwise_or(A_gray, B_gray)
xor_result_gray = cv2.bitwise_xor(A_gray, B_gray)
```

```python
# Mostrar las imágenes y sus resultados
plt.figure(figsize=(15, 6))

plt.subplot(2, 5, 1)
plt.imshow(cv2.cvtColor(A, cv2.COLOR_BGR2RGB))
plt.title("Imagen A")
plt.axis('off')
plt.subplot(2, 5, 2)
plt.imshow(cv2.cvtColor(B, cv2.COLOR_BGR2RGB))
plt.title("Imagen B")
plt.axis('off')
plt.subplot(2, 5, 3)
plt.imshow(cv2.cvtColor(and_result, cv2.COLOR_BGR2RGB))
plt.title("A and B")
plt.axis('off')
plt.subplot(2, 5, 4)
plt.imshow(cv2.cvtColor(or_result, cv2.COLOR_BGR2RGB))
plt.title("A or B")
plt.axis('off')
plt.subplot(2, 5, 5)
plt.imshow(cv2.cvtColor(xor_result, cv2.COLOR_BGR2RGB))
plt.title("A Xor B")
plt.axis('off')
plt.subplot(2, 5, 6)
plt.imshow(A_gray, cmap='gray')
plt.title("Imagen A")
plt.axis('off')
plt.subplot(2, 5, 7)
plt.imshow(B_gray, cmap='gray')
plt.title("Imagen B")
plt.axis('off')
plt.subplot(2, 5, 8)
plt.imshow(and_result_gray, cmap='gray')
plt.title("A and B")
plt.axis('off')
plt.subplot(2, 5, 9)
plt.imshow(or_result_gray, cmap='gray')
plt.title("A or B")
plt.axis('off')
plt.subplot(2, 5, 10)
plt.imshow(xor_result_gray, cmap='gray')
plt.title("A Xor B")
plt.axis('off')
plt.show()
```
