```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cargar la imagen
Img1 = cv2.imread('rgb.jpeg')
Img1 = cv2.cvtColor(Img1, cv2.COLOR_BGR2RGB)  # Convertir de BGR a RGB
Img1 = Img1.astype(np.float32) / 255.0  # Convertir a punto flotante y escalar al rango [0, 1]

# Mostrar la imagen original
plt.figure()
plt.imshow(Img1)
plt.title('Original')
plt.show()
```

```python
# Separar las componentes de color
Img1_red = Img1[:, :, 0]
Img1_green = Img1[:, :, 1]
Img1_blue = Img1[:, :, 2]

# Mostrar las componentes de color
plt.figure()
plt.subplot(1, 3, 1)
plt.imshow(Img1_red, cmap='gray')
plt.title('Componente Rojo')
plt.subplot(1, 3, 2)
plt.imshow(Img1_green, cmap='gray')
plt.title('Componente Verde')
plt.subplot(1, 3, 3)
plt.imshow(Img1_blue, cmap='gray')
plt.title('Componente Azul')
plt.show()
```

```python
# Sumar las componentes de color
ImgSumaComponentes = Img1_red + Img1_blue + Img1_green
plt.figure()
plt.imshow(ImgSumaComponentes, cmap='gray')
plt.title('Suma de componentes')
plt.show()
```

```python
# Crear imgenes con canales de color especficos
ImgRed = np.zeros_like(Img1)
ImgRed[:, :, 0] = Img1_red
ImgGreen = np.zeros_like(Img1)
ImgGreen[:, :, 1] = Img1_green
ImgBlue = np.zeros_like(Img1)
ImgBlue[:, :, 2] = Img1_blue

# Mostrar los canales de color
plt.figure()
plt.subplot(1, 3, 1)
plt.imshow(ImgRed)
plt.title('Canal Rojo')
plt.subplot(1, 3, 2)
plt.imshow(ImgGreen)
plt.title('Canal Verde')
plt.subplot(1, 3, 3)
plt.imshow(ImgBlue)
plt.title('Canal Azul')
plt.show()
```

```python
# Sumar los canales de color
ImgSumaCanales = ImgRed + ImgGreen + ImgBlue
plt.figure()
plt.imshow(ImgSumaCanales)
plt.title('Suma de Canales')
plt.show()
```
