```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cargar la imagen
Img1 = cv2.imread('image.png')
Img1 = cv2.cvtColor(Img1, cv2.COLOR_BGR2RGB)  # Convertir de BGR a RGB
Img1 = Img1.astype(np.float32) / 255.0  # Convertir a punto flotante y escalar al rango [0, 1]
```

```python
# Suma
plt.figure()
for i in np.arange(0, 1.2, 0.2):
    Img1_const = Img1 + i
    plt.subplot(1, 2, 1)
    plt.imshow(Img1_const)
    plt.title('Imagen original - Suma')
    plt.axis('off')
    plt.subplot(1, 2, 2)
    plt.hist(Img1_const.ravel(), bins=256, range=(0, 1), color='k')
    plt.title('Histograma')
    plt.axis('off')
    plt.pause(2)
plt.show()
```

```python
# Resta
plt.figure()
for i in np.arange(0, 1.2, 0.2):
    Img1_const = Img1 - i
    plt.subplot(1, 2, 1)
    plt.imshow(Img1_const)
    plt.title('Imagen original - Resta')
    plt.axis('off')
    plt.subplot(1, 2, 2)
    plt.hist(Img1_const.ravel(), bins=256, range=(0, 1), color='k')
    plt.title('Histograma')
    plt.axis('off')
    plt.pause(2)
plt.show()
```

```python
# Multiplicacin
plt.figure()
for i in np.arange(0, 1.1, 0.1):
    Img1_const = Img1 * i
    plt.subplot(1, 2, 1)
    plt.imshow(Img1_const)
    plt.title('Imagen original - Multiplicacin')
    plt.axis('off')
    plt.subplot(1, 2, 2)
    plt.hist(Img1_const.ravel(), bins=256, range=(0, 1), color='k')
    plt.title('Histograma')
    plt.axis('off')
    plt.pause(2)
plt.show()
```

```python
# Divisin
plt.figure()
for i in np.arange(0.1, 1.1, 0.1):
    Img1_const = Img1 / i
    plt.subplot(1, 2, 1)
    plt.imshow(Img1_const)
    plt.title('Imagen original - Divisin')
    plt.axis('off')
    plt.subplot(1, 2, 2)
    plt.hist(Img1_const.ravel(), bins=256, range=(0, 1), color='k')
    plt.title('Histograma')
    plt.axis('off')
    plt.pause(2)
plt.show()
```
