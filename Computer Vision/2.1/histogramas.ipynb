```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
Img_org = cv2.imread('simpson.jpg')
Img_org = cv2.cvtColor(Img_org, cv2.COLOR_BGR2RGB)  # Convertir de BGR a RGB
#Img_org = Img_org.astype(np.float32) / 255.0

plt.figure(figsize=(10,6))
plt.subplot(1, 2, 1)
plt.imshow(Img_org)
plt.title('Imagen original')
plt.axis('off')
plt.show()
```

```python
color = ('b','g','r')
for i, c in enumerate(color):
    hist = cv2.calcHist([Img_org], [i], None, [256], [0, 256])
    plt.plot(hist, color = c)
    plt.xlim([0,256])

plt.show()
```

```python
# Separar las componentes de color
Img1_red = Img_org[:, :, 0]
Img1_green = Img_org[:, :, 1]
Img1_blue = Img_org[:, :, 2]

# Mostrar las componentes de color y sus histogramas
plt.figure()
plt.subplot(2, 3, 1)
plt.imshow(Img1_red, cmap='gray')
plt.title('Componente Rojo')
plt.axis('off')
plt.subplot(2, 3, 2)
plt.imshow(Img1_green, cmap='gray')
plt.title('Componente Verde')
plt.axis('off')
plt.subplot(2, 3, 3)
plt.imshow(Img1_blue, cmap='gray')
plt.title('Componente Azul')
plt.axis('off')
plt.subplot(2, 3, 4)
plt.hist(Img1_red.ravel(), bins=256, range=(0.0, 1.0), color='r')
plt.title('Hist. del componente rojo')
plt.axis('off')
plt.subplot(2, 3, 5)
plt.hist(Img1_green.ravel(), bins=256, range=(0.0, 1.0), color='g')
plt.title('Hist. del componente verde')
plt.axis('off')
plt.subplot(2, 3, 6)
plt.hist(Img1_blue.ravel(), bins=256, range=(0.0, 1.0), color='b')
plt.title('Hist. del componente azul')
plt.axis('off')
plt.show()
```

```python
# Convertir la imagen a escala de grises
Img_gray = cv2.cvtColor(Img_org, cv2.COLOR_RGB2GRAY)

# Mostrar la imagen en escala de grises y su histograma
plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(Img_gray, cmap='gray')
plt.title('Imagen original en escala de grises')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.hist(Img_gray.ravel(), bins=256, range=(0.0, 1.0), color='k')
plt.title('Histograma en escala de grises')
plt.axis('off')
plt.show()
```

```python
# Crear imágenes con canales de color específicos
ImgRed = np.zeros_like(Img_org)
ImgRed[:, :, 0] = Img1_red
ImgGreen = np.zeros_like(Img_org)
ImgGreen[:, :, 1] = Img1_green
ImgBlue = np.zeros_like(Img_org)
ImgBlue[:, :, 2] = Img1_blue

# Mostrar los canales de color y sus histogramas
plt.figure()
plt.subplot(2, 3, 1)
plt.imshow(ImgRed)
plt.title('Canal Rojo')
plt.axis('off')
plt.subplot(2, 3, 2)
plt.imshow(ImgGreen)
plt.title('Canal Verde')
plt.axis('off')
plt.subplot(2, 3, 3)
plt.imshow(ImgBlue)
plt.title('Canal Azul')
plt.axis('off')
plt.subplot(2, 3, 4)
plt.hist(ImgRed[:, :, 0].ravel(), bins=256, range=(0.0, 1.0), color='r')
plt.title('Hist. del canal rojo')
plt.axis('off')
plt.subplot(2, 3, 5)
plt.hist(ImgGreen[:, :, 1].ravel(), bins=256, range=(0.0, 1.0), color='g')
plt.title('Hist. del canal verde')
plt.axis('off')
plt.subplot(2, 3, 6)
plt.hist(ImgBlue[:, :, 2].ravel(), bins=256, range=(0.0, 1.0), color='b')
plt.title('Hist. del canal azul')
plt.axis('off')
plt.show()
```
