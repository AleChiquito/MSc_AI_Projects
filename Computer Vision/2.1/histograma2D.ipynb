```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cargar la imagen
imagen = cv2.imread('simpson.jpg')

# Dividir la imagen en canales RGB
canal_R = imagen[:, :, 0]
canal_G = imagen[:, :, 1]
canal_B = imagen[:, :, 2]
```

```python
# Crear una figura y varios subgrficos para los histogramas
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Calcular y mostrar el histograma para el canal RG
hist_RG, bins_RG = np.histogram(np.concatenate((canal_R.flatten(), canal_G.flatten())), bins=256, range=(0, 255))
axs[0].plot(hist_RG, color='purple')
axs[0].set_title('Histograma de RG')
axs[0].set_xlabel('Intensidad')
axs[0].set_ylabel('Frecuencia')

# Calcular y mostrar el histograma para el canal RB
hist_RB, bins_RB = np.histogram(np.concatenate((canal_R.flatten(), canal_B.flatten())), bins=256, range=(0, 255))
axs[1].plot(hist_RB, color='green')
axs[1].set_title('Histograma de RB')
axs[1].set_xlabel('Intensidad')
axs[1].set_ylabel('Frecuencia')

# Calcular y mostrar el histograma para el canal GB
hist_GB, bins_GB = np.histogram(np.concatenate((canal_G.flatten(), canal_B.flatten())), bins=256, range=(0, 255))
axs[2].plot(hist_GB, color='blue')
axs[2].set_title('Histograma de GB')
axs[2].set_xlabel('Intensidad')
axs[2].set_ylabel('Frecuencia')

# Mostrar los histogramas
plt.tight_layout()
plt.show()
```
