```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = cv2.imread('chica_mini.png')
A = cv2.cvtColor(A, cv2.COLOR_BGR2RGB)
A = A.astype(np.float32) / 255.0
```

```python
# Definir las máscaras
mask = np.array([[-1, -1, -1],
                 [-1,  8, -1],
                 [-1, -1, -1]], dtype=np.float32)

a = 5
mask2 = np.array([[0, -a, 0],
                  [-a, (4*a)+1, -a],
                  [0, -a, 0]], dtype=np.float32)
```

```python
# Redimensionar la imagen
R1 = cv2.resize(A, None, fx=10, fy=10, interpolation=cv2.INTER_NEAREST)
R2 = cv2.resize(A, None, fx=10, fy=10, interpolation=cv2.INTER_LINEAR)
R3 = cv2.resize(A, None, fx=10, fy=10, interpolation=cv2.INTER_CUBIC)

# Aplicar convolución a las imágenes redimensionadas
R1P = cv2.filter2D(R1, -1, mask2)
R2P = cv2.filter2D(R2, -1, mask2)
R3P = cv2.filter2D(R3, -1, mask2)
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(3, 2, 1)
plt.imshow(R1)
plt.title("Imagen ampliada x10 (Nearest Neighbor)")
plt.axis('off')

plt.subplot(3, 2, 2)
plt.imshow(R1P)
plt.title("Imagen ampliada - Vecinos (Nearest Neighbor)")
plt.axis('off')

plt.subplot(3, 2, 3)
plt.imshow(R2)
plt.title("Imagen ampliada x10 (Bilinear)")
plt.axis('off')

plt.subplot(3, 2, 4)
plt.imshow(R2P)
plt.title("Imagen ampliada - Bilineal (Bilinear)")
plt.axis('off')

plt.subplot(3, 2, 5)
plt.imshow(R3)
plt.title("Imagen ampliada x10 (Bicubic)")
plt.axis('off')

plt.subplot(3, 2, 6)
plt.imshow(R3P)
plt.title("Imagen ampliada - Bicúbica (Bicubic)")
plt.axis('off')

plt.show()
```
