```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cargar la imagen
A = cv2.imread('chico.jpg')
# Convertir la imagen a punto flotante
A = A.astype(np.float32) / 255.0

# Aplicar umbralizaci√≥n
umbral_1 = 42 / 255
umbral_2 = 180 / 255

R1 = np.where(cv2.cvtColor(A, cv2.COLOR_BGR2GRAY) < umbral_1, 0, 1)
R2 = np.where(cv2.cvtColor(A, cv2.COLOR_BGR2GRAY) < umbral_2, 0, 1)
```

```python
# Mostrar las im√°genes
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(A, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(R1, cmap='gray')
plt.title('Umbral = 42')
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(R2, cmap='gray')
plt.title('Umbral = 180')
plt.axis('off')
plt.show()

```
