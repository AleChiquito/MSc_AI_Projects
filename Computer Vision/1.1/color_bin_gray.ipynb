```python
import matplotlib.pyplot as plt
from PIL import Image
```

```python
def convert_to_binary(image_path, threshold=128):
    # Open the image
    img = Image.open(image_path)
    # Convert to grayscale
    gray_img = img.convert('L')
    # Convert to binary
    bin_img = gray_img.point(lambda p: p > threshold and 255)
    return bin_img

def convert_to_grayscale(image_path):
    # Open the image
    img = Image.open(image_path)
    # Convert to grayscale
    gray_img = img.convert('L')
    return gray_img
```

```python
# Example usage:
input_image_path = "simpson.jpg"

# Convert to binary
binary_image = convert_to_binary(input_image_path)

# Convert to grayscale
grayscale_image = convert_to_grayscale(input_image_path)
```

```python
def display_images(images, titles):
    num_images = len(images)
    fig, axes = plt.subplots(1, num_images, figsize=(15, 5))
    
    for i in range(num_images):
        axes[i].imshow(images[i], cmap='gray')
        axes[i].set_title(titles[i])
        axes[i].axis('off')
    
    plt.show()

# Example usage:
input_image_path = "simpson.jpg"

# Convert to binary
binary_image = convert_to_binary(input_image_path)

# Convert to grayscale
grayscale_image = convert_to_grayscale(input_image_path)

# Display all three images
display_images([Image.open(input_image_path), binary_image, grayscale_image],
               ["Original Image", "Binary Image", "Grayscale Image"])
```
