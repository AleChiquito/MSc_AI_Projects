```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cargar la imagen
Img1 = cv2.imread('car_dark.jpeg')
Img1 = cv2.cvtColor(Img1, cv2.COLOR_BGR2RGB)  # Convertir de BGR a RGB
Img1 = Img1.astype(np.float32) / 255.0  # Convertir a punto flotante y escalar al rango [0, 1]
```

```python
# Mostrar la imagen original y su histograma
plt.figure()
plt.subplot(2, 2, 1)
plt.imshow(Img1)
plt.title('Imagen original')
plt.axis('off')
plt.subplot(2, 2, 2)
plt.hist(Img1.ravel(), bins=256, range=(0, 1), color='k')
plt.title('Histograma')
plt.axis('off')
```

```python
# Normalizar la imagen
m = 0
M = 0.2
ImgRes = ((Img1 - m) / (M - m))
ImgRes = np.clip(ImgRes, 0, 1)  # Asegurar que los valores estén en el rango [0, 1]
```

```python
# Mostrar la imagen resultante y su histograma
plt.subplot(2, 2, 3)
plt.imshow(ImgRes)
plt.title('Imagen resultante')
plt.axis('off')
plt.subplot(2, 2, 4)
plt.hist(ImgRes.ravel(), bins=256, range=(0, 1), color='k')
plt.title('Histograma resultante')
plt.axis('off')

# Guardar la imagen resultante
#cv2.imwrite("Car_light.jpeg", cv2.cvtColor((ImgRes * 255).astype(np.uint8), cv2.COLOR_RGB2BGR))

plt.show()
```
