```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
imagen = cv2.imread('perritos.jpg')

if len(imagen.shape) == 3:
    imagen_gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
else:
    imagen_gris = imagen

# Ecualizar el histograma de la imagen
imagen_ecualizada = cv2.equalizeHist(imagen_gris)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(imagen_gris, cmap='gray')
plt.title('Imagen original')
plt.axis('off')
plt.subplot(2, 2, 2)
plt.hist(imagen_gris.ravel(), bins=256, range=(0, 256))
plt.title('Histograma')

plt.subplot(2, 2, 3)
plt.imshow(imagen_ecualizada, cmap='gray')
plt.title('Imagen resultante')
plt.axis('off')
plt.subplot(2, 2, 4)
plt.hist(imagen_ecualizada.ravel(), bins=256, range=(0, 256))
plt.title('Histograma resultante')


plt.show()
```
