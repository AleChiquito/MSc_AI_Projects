```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-14, 8, 0.1)
y = (x + 3) ** 2

plt.plot(x,y)
plt.show()
```

```python
def grad_des(w, lr, epocas):
    resultados = []
    for i in range(epocas):
        w =  w - lr * (2*w + 6)
        resultados.append(w)
    return resultados
```

```python
resultados = grad_des(7, 0.1, 200)
```

```python
resultados
```

```python
plt.plot(resultados)
plt.show()
```
